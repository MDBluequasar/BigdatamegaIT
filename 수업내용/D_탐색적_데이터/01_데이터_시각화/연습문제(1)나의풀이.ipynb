```python
plt.figure()
plt.title('비트코인의 날짜별 종가와 시가의 변화')

plt.xlabel('날짜')
plt.ylabel('가격')

plt.plot(xlabels,   df['시가'], color="red", linestyle=':', linewidth=1)
plt.plot(xlabels,   df['종가'], color="green", linestyle='--', linewidth=1)


plt.grid()
plt.legend()
plt.show()
plt.close()
```
