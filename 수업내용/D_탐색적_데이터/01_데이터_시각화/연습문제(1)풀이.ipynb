```python
df = read_excel('C:\\Users\\User\\Desktop\\궁둥박사\\BigdatamegaIT\\수업내용\\D_탐색적_데이터\\01_데이터_시각화//traffic_acc.xlsx')
df
```

```python
plt.rcParams["figure.figsize"] = (15, 7)

plt.figure()
plt.scatter(df['발생건수'], df['부상자수'])
plt.xlabel('발생건수')
plt.ylabel('부상자수')
plt.grid()
plt.show()
plt.close()
```

```python
# 추가 내용 - numpy를 활용한 추세선 구하기(회귀분석)

z = np.polyfit(df['발생건수'], df['부상자수'], 1)
print('기울기(가중치)=', z[0])
print('절편(편향)=', z[1])
```

```python
# 발생건수에 대한 부상자 수 예측
print("f(x) = %f * x + %f" % (z[0], z[1]))
```

```python
f = np.poly1d(z)
f
```

```python
f(1)
```

```python
f(df['발생건수'])
```

```python
plt.rcParams["figure.figsize"] = (15, 7)

plt.figure()
plt.scatter(df['발생건수'], df['부상자수'])
plt.plot(df['발생건수'], f(df['발생건수']), color = '#ff0000') # 선 추가

plt.xlabel("발생건수")
plt.ylabel("부상자수")
plt.grid()
plt.show()
plt.close()
```
