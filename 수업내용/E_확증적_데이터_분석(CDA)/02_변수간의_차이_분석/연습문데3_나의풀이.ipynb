# 문제 1

유아들을 대상으로 세 가지 읽는 방법을 비교,실험하여 다음과 같은 독서평가 점수자료를 얻었다.

세 방법에 통계적으로 유의한 차이가 있는지 검정하라

```python
df = read_excel('https://data.hossam.kr/E02/reading.xlsx')
df
```

```python
df2 = df.copy()
df2['독서방법'] = df2['독서방법'].map({'B': 1, 'D': 2, 'S': 3})
df2
```

```python
all_test(df2['독서방법'], df2['점수'])
```

```python
model = ols('점수 ~ C(독서방법)', data = df2).fit()
anova_lm(model)
```

#### 결과 해석
F(8, 5) = 2.657563 / p-value > 0.05 이상이므로 독서 방법에 따라 통계적으로 유의한 차이가 없다.
