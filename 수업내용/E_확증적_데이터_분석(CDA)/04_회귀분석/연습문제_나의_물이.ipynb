```python
from pandas import read_excel
from statsmodels.formula.api import ols
from matplotlib import pyplot as plt
import seaborn as sb
import sys
import os

sys.path.append(os.path.dirname(os.path.dirname(os.getcwd())))
from helper import pearson_r
```

# 문제 1.
아래의 데이터는 콩의 수분 함유량에 따른 강도를 조사한 데이터이다. 위 데이터를 분석하라.

```python
bf = read_excel('https://data.hossam.kr/E04/beans.xlsx')
bf
```

```python
plt.rcParams["font.family"] = 'AppleGothic' if sys.platform == 'darwin' else 'Malgun Gothic'
plt.rcParams["font.size"] = 12
plt.rcParams["figure.figsize"] = (10, 5)
plt.rcParams["axes.unicode_minus"] = False
```

```python
plt.rcParams['figure.figsize'] = (10, 5)
sb.regplot(data=bf, x='수분함유량', y='강도')
plt.grid()
plt.show()
plt.close()
```

```python
pearson_r(bf)
```

```python
model = ols("강도 ~ 수분함유량", data = bf).fit().summary() 
model
```

# 문제 2
아래의 데이터는 10년간 에어컨의 예약대수에 따른 실 판매대수를 조사한 가상의 데이터이다.

- 1. 위 데이터를 분석하고 결과를 보고하시오.
- 2. 기록적인 폭염으로 올해 에어컨 예약 대수는 150대였다. 실 판매대수를 예측하시오.


```python
af = read_excel('https://data.hossam.kr/E04/aircon.xlsx')
af
```

```python
plt.rcParams['figure.figsize'] = (10, 5)
sb.regplot(data=af, x='예약대수', y='판매대수')
plt.grid()
plt.show()
plt.close()
```

```python
pearson_r(af)
```

```python
model = ols("예약대수 ~ 판매대수", data = af).fit().summary() 
model
```

# 문제 3.

아래의 데이터는 부모의 키에 따른 자식의 키를 조사한 데이터이다.

  - 1. 위 데이터를 분석하고 결과를 보고하시오.
  - 2. 당신이 자식을 낳는다면 키가 어느정도일지 예측해 보시오.


```python
hf = read_excel('https://data.hossam.kr/E04/galton.xlsx')
hf
```

```python
plt.rcParams['figure.figsize'] = (10, 5)
sb.regplot(data = hf, x= 'child', y = 'parent')
plt.grid()
plt.show()
plt.close()
```

```python
pearson_r(hf)
```

```python
model = ols("child ~ parent", data = hf).fit().summary() 
model
```
